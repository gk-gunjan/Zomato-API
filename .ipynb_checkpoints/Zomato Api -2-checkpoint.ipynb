## 2.1.4 User Rating v/s Restaurant serving some specific cuisines

```python
file = pd.read_csv('zomato.csv', encoding = 'iso-8859-1')
df = file.copy()

# considering only INDIAN Restraunts - 
df = df[df['Country Code'] == 1]

# creating dict containing names of cuisines and their count.
cuisine=dict()
for i in df["Cuisines"]:
    a=i.split(',')
    for j in a:
        j=j.strip()
        cuisine[j]=cuisine.get(j, 0) + 1

# sorting dictionary according to the count of cuisines and fetching top 10 cuisines.
popular=sorted(cuisine.items(),key=lambda kv:kv[1], reverse=True)[:5] 

# fetching names of top 10 cuisines.
popular_cuisines=[]
for i in popular:
    popular_cuisines.append(i[0])

plt.figure(num=None, figsize=(16, 10))

# fetching ratings of the particular cuisine of particular restaurant.
for i in popular_cuisines:
    rating=[]
    for j, k in zip(df.Cuisines, df['Aggregate rating']):
        if i in j:
            rating.append(k)
            
# plotting histogram for each cuisine .
    plt.hist(rating, edgecolor='black', bins=[0, 1, 2, 3, 4, 5])
    plt.xlabel('Rating--->')
    plt.ylabel('Number of restaurants--->', size = 12)
    plt.title('variation of aggregate rating with restaurants serving some specific cuisines(top 5)', size = 12)
plt.grid()
plt.legend(labels=popular_cuisines)
plt.show()
```

## 2.2 Find the weighted restaurant rating of each locality and find out the top localities with more weighted restaurant rating?

#### Weighted Restaurant Rating=Σ (number of votes * rating) / Σ (number of votes) .


```python
#Creating a dictionary of locality:[votes*ratings,votes] which will be further used in calculating weighted restaurant rating
df.dropna(subset=['Votes'],inplace=True)
df.dropna(subset=['Locality'],inplace=True)
df.dropna(subset=['Aggregate rating'],inplace=True)
df.dropna(subset=['Restaurant Name'],inplace=True)

d={}
for i in df.index:
    votes = df['Votes'][i]
    loc = df['Locality'][i]
    ratings = df['Aggregate rating'][i]
    rest = df['Restaurant Name'][i]
    
    if loc not in d:
        d[loc] = [votes*ratings,votes]
    else:
        d[loc][0] += votes*ratings
        d[loc][1] += votes

#calculating Weighted Restaurant Rating for each location
for i in d.items():
    
    x=i[1][0]
    y=i[1][1]
    w=x/y
    d[i[0]]=w
```

```python
print(d)
```

```python
print("Top localities with more weighted restaurant rating are:")
print("")
sort_loc = sorted(d.items(), key=lambda x: x[1], reverse=True)[:8]
for i,j in sort_loc:
    print(i, j)
```

## 3.5 Number of Restaurants present in the city of India 

### (keeping the weighted restaurant rating of the city in a bubble)

```python
file = pd.read_csv('zomato.csv', encoding = 'iso-8859-1')
df = file.copy()
df = df[df['Country Code'] == 1]
d={}
for i in df.index:
    votes = df['Votes'][i]
    loc = df['City'][i]
    ratings = df['Aggregate rating'][i]
    
    if loc not in d:
        d[loc] = [votes*ratings,votes,1]
    else:
        d[loc][0] += votes*ratings
        d[loc][1] += votes
        d[loc][2] +=1


dic={}
cities=[]
weig_rat=[]
no_res=[]
for i in d.items():
    if i[0] not in dic:
        x=i[1][0]
        y=i[1][1]
        w=x/y
        weig_rat.append(w)
        cities.append(i[0])
        no_res.append(i[1][2])
```

```python
plt.figure(figsize=(12,15))
plt.scatter(no_res,cities,s=weig_rat,c='r')
plt.show()
```
